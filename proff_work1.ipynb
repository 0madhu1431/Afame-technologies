### Visualizing Differences using PCA

To visualize the differences between `features_df1` and `features_df2`, I'll use Principal Component Analysis (PCA) to reduce their dimensionality. PCA transforms the data into a new set of orthogonal variables called principal components, which capture the maximum variance in the data. By plotting the first two principal components, we can observe any separation or clustering between the two feature sets.

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np

# Combine the features for PCA
all_features = np.vstack((features_df1, features_df2))

# Apply PCA to reduce to 2 components
pca = PCA(n_components=2)
principal_components = pca.fit_transform(all_features)

# Separate the principal components back into df1 and df2 components
pca_df1 = principal_components[:len(features_df1)]
pca_df2 = principal_components[len(features_df1):]

# Plot the results
plt.figure(figsize=(10, 7))
plt.scatter(pca_df1[:, 0], pca_df1[:, 1], alpha=0.7, label='Patient 1 Features', color='blue')
plt.scatter(pca_df2[:, 0], pca_df2[:, 1], alpha=0.7, label='Patient 2 Features', color='red')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Patient 1 vs Patient 2 Features')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.fft import fft
import numpy as np

N = len(signal)
fft_vals = np.abs(fft(signal))

# Define a default sampling frequency. Adjust if you know the actual fs.
fs = 1 # Assuming a sampling frequency of 1 unit

freqs = np.fft.fftfreq(N, 1/fs)

# take only positive frequencies
positive_freqs = freqs[:N//2]
positive_fft = fft_vals[:N//2]

dominant_freq_patient1 = positive_freqs[np.argmax(positive_fft)]
print("Dominant Frequency for Patient 1:", dominant_freq_patient1)


# Calculate dominant frequency for signal1 (Patient 2)
N1 = len(signal1)
fft_vals1 = np.abs(fft(signal1))
freqs1 = np.fft.fftfreq(N1, 1/fs)

positive_freqs1 = freqs1[:N1//2]
positive_fft1 = fft_vals1[:N1//2]

dominant_freq_patient2 = positive_freqs1[np.argmax(positive_fft1)]
print("Dominant Frequency for Patient 2:", dominant_freq_patient2)
```

**BAND POWERS**

```python
def bandpower(freqs, psd, low, high):
    idx = np.logical_and(freqs >= low, freqs <= high)
    return np.trapezoid(psd[idx], freqs[idx])

print("Band powers for Patient 1:")
delta_p1 = bandpower(freqs_p1, psd_p1, 0.5, 4)
theta_p1 = bandpower(freqs_p1, psd_p1, 4, 8)
alpha_p1 = bandpower(freqs_p1, psd_p1, 8, 13)
beta_p1  = bandpower(freqs_p1, psd_p1, 13, 30)

print("Delta:", delta_p1)
print("Theta:", theta_p1)
print("Alpha:", alpha_p1)
print("Beta:", beta_p1)

print("\nBand powers for Patient 2:")
delta_p2 = bandpower(freqs_p2, psd_p2, 0.5, 4)
theta_p2 = bandpower(freqs_p2, psd_p2, 4, 8)
alpha_p2 = bandpower(freqs_p2, psd_p2, 8, 13)
beta_p2  = bandpower(freqs_p2, psd_p2, 13, 30)

print("Delta:", delta_p2)
print("Theta:", theta_p2)
print("Alpha:", alpha_p2)
print("Beta:", beta_p2)
```

**FINDING PATTERNS**

```python
import numpy as np

def segment_signal(signal, segment_size=20):
    segments = []
    for i in range(0, len(signal), segment_size):
        segments.append(signal[i:i+segment_size])
    return segments

segments1 = segment_signal(signal)
segments2 = segment_signal(signal1)
print(segments1)
print(segments2)
```

**EXTRACTING FOR EACH SEGMENT**

```python
def extract_features(segment):
    mean = np.mean(segment)
    std = np.std(segment)
    energy = np.sum(segment**2)
    zcr = np.sum(np.diff(np.sign(segment)) != 0)

    return [mean, std, energy, zcr]

features1 = [extract_features(seg) for seg in segments1]
features2 = [extract_features(seg) for seg in segments2]
print(features1)
print(features2)
```

```python
import pandas as pd

df_feat1 = pd.DataFrame(features1, columns=["Mean","Std","Energy","ZCR"])
df_feat2 = pd.DataFrame(features2, columns=["Mean","Std","Energy","ZCR"])
print(df_feat1)
print(df_feat2)
```
